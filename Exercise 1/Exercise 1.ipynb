```python
#Task 4
county = {} #Dictionary to store county populations
with open("norway_municipalities_2017.csv") as municipalities:
    municipalities.readline() #Skips first line with text
    for line in municipalities: #Reads every line with in the file
        #Separates the lines into a list for later use. 
        #The indexes correspond to the following: 0 is the municipalitiy, 1 is the county, 2 is the population
        splitLine = line.split(",") 
        if splitLine[1] in county.keys(): #Checks if the county has been added to  the dictionary
            #Adds the population value to the currently stored county population
            county[splitLine[1]] = county[splitLine[1]] + int(splitLine[2]) 
        else:
            county[splitLine[1]] = int(splitLine[2]) #Adds the county to  the dict if it has not been added

print("-"*17+"Alphabetically sorted"+"-"*17)
#alphabetically sorted
sortedCountiesByName = sorted(county.keys()) #Sorts the counties alphabetically 
for keys in sortedCountiesByName: 
    print(f"County: {keys:<25} Population: {county[keys]}") #Prints a table sorted alphabetically
print("-"*19+"Population Sorted"+"-"*19)

#Population sorted
sortedCountiesByPopulation = sorted(county.values()) #Sorts by the population
sortedCountiesByPopulation.reverse() #Reverses list order such that it is in descending order

for values in sortedCountiesByPopulation:
    #Prints a table sorted by population. County name uses the values of the dict to find the keys
    print(f"County: {list(county.keys())[list(county.values()).index(values)]:<25} Population: {values}") 
print("-"*55)
```
